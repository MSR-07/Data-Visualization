```python
#import Libraries
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Data

df = pd.read_csv(".\Data_Set\kidney_disease.csv")

#Print Frist Five Rows,
#df.head()
df.head(400).T
```

```python
df[['htn','dm','cad','pe','ane']] = df[['htn','dm','cad','pe','ane']].replace(to_replace={'yes':1,'no':0})
```

```python
df[['rbc','pc']] = df[['rbc','pc']].replace(to_replace={'normal':0,'abnormal':1})
```

```python
df[['ba','pcc']] = df[['ba','pcc']].replace(to_replace={'present':1})
```

```python
df[['appet']] = df[['appet']].replace(to_replace={'good':1,'poor':0})
```

```python
df["classification"] = [1 if i == "ckd" else 0 for i in df["classification"]]
```

```python
((df.isnull().sum()/df.shape[0])*100).sort_values(ascending=False).plot(kind='bar', figsize=(10,10))
```

```python
# blood pressure graph
sns.catplot(data=df, x='bp', kind= 'count',height=6,aspect=2)
plt.xlabel('Chronic Kidney Disease')
plt.title("blood pressure graph",fontsize=15)
plt.show()
```

```python
#sugar-frequency graph
sns.catplot(data=df, x='su', kind= 'count',height=6,aspect=2)
plt.xlabel('Chronic Kidney Disease')
plt.title("sugar-frequency graph",fontsize=15)
plt.show()
```

```python
sns.catplot(data=df, x='age', kind= 'count',aspect=5)
plt.xlabel('Chronic Kidney Disease')
plt.title("packed cell volume grahp",fontsize=15)
plt.show()
```
